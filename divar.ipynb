```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from unidecode import unidecode
```

```python
divar=pd.read_csv('divar.csv')
```

```python
divar
```

```python
divar.drop(columns=["real_estate_agencey", "units_per_floor", "document", "building_direction", "unit_status"], inplace=True)
```

```python
divar['facilities'].fillna(',',inplace=True)
divar['facilities'] = divar['facilities'].str.replace("[", "")
divar['facilities'] = divar['facilities'].str.replace("]", "")
divar['facilities'] = divar['facilities'].str.replace("'", "")
```

```python
exploded_df = divar['facilities'].map(lambda x:x.split(",")).explode()
one_hot_encoded = pd.get_dummies(exploded_df , dtype="int")
result_df = one_hot_encoded.groupby(one_hot_encoded.index).sum()
result_df.drop(columns="",  inplace=True)
result_df.drop(columns=['facilities'], inplace=True)
```

```python
result_df = result_df.loc[:, ["بالکن", "بالکن ندارد", "انباری",  "انباری ندارد" ,"آسانسور" ,"آسانسور ندارد" ,"پارکینگ" ,"پارکینگ ندارد"]]
```

```python
result_df.rename(columns={"پارکینگ": "has_parking",
                  "پارکینگ ندارد" : "no_parking",
                  "آسانسور": "has_elevator",
                  "آسانسور ندارد": "no_elevator",
                  "انباری": "has_warehouse",
                  "انباری ندارد": "no_warehouse",
                  "بالکن": "has_balcony",
                  "بالکن ندارد": "no_balcony"},
                inplace=True)
```

```python
merged_df = divar.merge(result_df, left_index=True, right_index=True)
```

```python
residental=merged_df[merged_df['real_estate_type']=='apartment-sell'].copy()
```

```python
residental['price'] = residental['price'].str.replace('٬','')
residental['price'] = residental['price'].str.replace('تومان','')
residental['price']=residental['price'].astype(str).apply(unidecode)
residental.drop(residental[residental['price'] == 'twfqy'].index,inplace=True)
residental["price"] = residental["price"].astype("float")
```

```python
residental['price_per_meter'] = residental['price_per_meter'].str.replace('٬','')
residental['price_per_meter'] = residental['price_per_meter'].str.replace('تومان','')
residental['price_per_meter']=residental['price_per_meter'].astype(str).apply(unidecode)
residental["price_per_meter"] = residental["price_per_meter"].astype("float")
```

```python
residental['floor']=residental['floor'].str.replace('زیرهمکف','-1')
residental['floor']=residental['floor'].str.replace('همکف','0')
residental['floor']=residental['floor'].str.replace('بیشتر از','')
residental['floor']=residental['floor'].str.replace('از','|')
residental['floor']=residental['floor'].astype(str).apply(unidecode)
residental['floor']=residental['floor'].str.split("|").apply(lambda x:x[0])
residental['floor']=residental['floor'].str.replace('30+', '30')
residental['floor']=residental['floor'].str.replace('+30', '30')
residental.drop(residental[residental["floor"] == 'nan'].index,inplace=True)
residental["floor"] = residental["floor"].astype("int")
```

```python
residental['real_estate_area']=residental['real_estate_area'].astype(str).apply(unidecode)
residental["real_estate_area"] = residental["real_estate_area"].astype('float')
```

```python
residental['real_estate_age']=residental['real_estate_age'].astype(str).apply(unidecode)
residental['real_estate_age']=residental['real_estate_age'].str.replace('qbl z 1370', '1370')
residental["real_estate_age"] = 1402 - (residental["real_estate_age"].astype('int'))
```

```python
residental['total_rooms']=residental['total_rooms'].astype(str).apply(unidecode)
residental['total_rooms']=residental['total_rooms'].str.replace('bdwn tq', '0')
residental['total_rooms']=residental['total_rooms'].str.replace('+4', '5')
residental["total_rooms"] = residental["total_rooms"].astype('int')
```

```python
special_word1 = 'استخر'
special_word2 = 'روف گاردن'
special_word3 = 'لابی'
special_word4 = 'لابی من'
special_word5 = 'آنتن مرکزی'
special_word7 = 'بالکن'
special_word8 = 'سونا'
special_word9 = 'جکوزی'
special_word10 = 'باشگاه'
special_word11 = 'جاروبرقی مرکزی'
special_word12 = 'سرایدار'
special_word13 = 'نگهبان'
special_word14 = 'مستر'
special_word15 = 'سالن اجتماعات'
special_word16 = 'پارکینگ'
special_word17 = 'آسانسور'
special_word19 = 'انباری'

residental['pool'] = residental[['description1','description2','description3','title']].apply(lambda row: row.astype(str).str.contains(special_word1, regex=True).any(), axis=1).astype(int)
residental['roof_garden'] = residental[['description1','description2','description3','title']].apply(lambda row: row.astype(str).str.contains(special_word2, regex=True).any(), axis=1).astype(int)
residental['lobby'] = residental[['description1','description2','description3','title']].apply(lambda row: row.astype(str).str.contains(special_word3, regex=True).any(), axis=1).astype(int)
residental['lobby_man'] = residental[['description1','description2','description3','title']].apply(lambda row: row.astype(str).str.contains(special_word4, regex=True).any(), axis=1).astype(int)
residental['central_anthena'] = residental[['description1','description2','description3','title']].apply(lambda row: row.astype(str).str.contains(special_word5, regex=True).any(), axis=1).astype(int)
residental['balkony'] = residental[['description1','description2','description3','title']].apply(lambda row: row.astype(str).str.contains(special_word7, regex=True).any(), axis=1).astype(int)
residental['sauna'] = residental[['description1','description2','description3','title']].apply(lambda row: row.astype(str).str.contains(special_word8, regex=True).any(), axis=1).astype(int)
residental['jacuzzi'] = residental[['description1','description2','description3','title']].apply(lambda row: row.astype(str).str.contains(special_word9, regex=True).any(), axis=1).astype(int)
residental['gym'] = residental[['description1','description2','description3','title']].apply(lambda row: row.astype(str).str.contains(special_word10, regex=True).any(), axis=1).astype(int)
residental['central_Vacuume_cleaner'] = residental[['description1','description2','description3','title']].apply(lambda row: row.astype(str).str.contains(special_word11, regex=True).any(), axis=1).astype(int)
residental['janitor'] = residental[['description1','description2','description3','title']].apply(lambda row: row.astype(str).str.contains(special_word12, regex=True).any(), axis=1).astype(int)
residental['Guard'] = residental[['description1','description2','description3','title']].apply(lambda row: row.astype(str).str.contains(special_word13, regex=True).any(), axis=1).astype(int)
residental['master_room'] = residental[['description1','description2','description3','title']].apply(lambda row: row.astype(str).str.contains(special_word14, regex=True).any(), axis=1).astype(int)
residental['conference_hall']= residental[['description1','description2','description3','title']].apply(lambda row: row.astype(str).str.contains(special_word15, regex=True).any(), axis=1).astype(int)
residental['parking_d']= residental[['description1','description2','description3','title']].apply(lambda row: row.astype(str).str.contains(special_word16, regex=True).any(), axis=1).astype(int)
residental['elevator_d'] = residental[['description1','description2','description3','title']].apply(lambda row: row.astype(str).str.contains(special_word17, regex=True).any(), axis=1).astype(int)
residental['warehouse_d']= residental[['description1','description2','description3','title']].apply(lambda row: row.astype(str).str.contains(special_word19, regex=True).any(), axis=1).astype(int)
```

```python
def func(row):
    if row.parking is np.nan:
        if (row.has_parking == 0) and (row.no_parking == 0):
            row.has_parking = np.nan
            row.no_parking = np.nan
            
        if row.has_parking == 1:
            row.parking = True
        elif row.has_parking == 0:
            row.parking = False
        elif row.parking_d == 1:
            row.parking = True
            
    if row.elevator is np.nan:
        if (row.has_elevator == 0) and (row.no_elevator == 0):
                row.has_elevator = np.nan
                row.no_elevator = np.nan
            
        if row.has_elevator == 1:
            row.elevator = True
        elif row.has_elevator == 0:
            row.elevator = False
        elif row.elevator_d == 1:
            row.elevator = True
            

    if row.warehouse is np.nan:
        if (row.has_warehouse == 0) and (row.no_warehouse == 0):
                row.has_warehouse = np.nan
                row.no_warehouse = np.nan
        if row.has_warehouse == 1:
            row.warehouse = True
        elif row.has_warehouse == 0:
            row.warehouse = False
        elif row.warehouse_d == 1:
            row.warehouse = True

    if (row.has_balcony == 0) and (row.no_balcony == 0):
        row.has_balcony = np.nan
        row.no_balcony = np.nan
        
    if row.has_balcony is np.nan:
        if row.balkony == 1:
            row.has_balcony = 1
    

    return row


residental = residental.apply(func, axis=1)
```
